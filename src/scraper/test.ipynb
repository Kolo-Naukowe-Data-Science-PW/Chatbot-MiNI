```python
from trafilatura import extract, extract_metadata, fetch_url, html2txt


def fetch(url: str) -> str:
    html = fetch_url(url)
    if not html:
        return ""

    return extract(
        html,
        url=url,
        favor_recall=True,
        include_tables=False,
        include_comments=False,
        deduplicate=False,
        with_metadata=True,
    ) or ""

def extract_text(url: str) -> str:
    downloaded = fetch_url(url)
    result = html2txt(downloaded, clean=True)
    metadata = extract_metadata(downloaded)
    return result, metadata

if __name__ == "__main__":
    urls = [
        "https://www.pw.edu.pl/studia/harmonogram-roku-akademickiego",
        "https://ww2.mini.pw.edu.pl/studia/inzynierskie-i-licencjackie/matematyka-i-analiza-danych/",
        "https://ww2.mini.pw.edu.pl/studia/inzynierskie-i-licencjackie/rankingi-i-opinie/",
        "https://ww2.mini.pw.edu.pl/",
        "https://ww2.mini.pw.edu.pl/studia/plany-zajec-i-procedury/plan-sesji/",
        "https://ww2.mini.pw.edu.pl/studia/plany-zajec-i-procedury/plany-zajec/",
        "https://ww2.mini.pw.edu.pl/studia/dziekanat/kontakt-i-dyzury/",
        "https://ww2.mini.pw.edu.pl/studia/dziekanat/informacje-dziekanatu/",
        "https://ww2.mini.pw.edu.pl/studia/dziekanat/dla-dyplomantow-i-promotorow/"

    ]

    for i, url in enumerate(urls):
        text = fetch(url)
        f = open(f"output{i}.txt", "w", encoding="utf-8")
        f.write(text)
        f.close()

        
```
